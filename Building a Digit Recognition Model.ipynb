```python
digits = datasets.load_digits()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.tick_params(
    axis='both', which='both', bottom=False, top=False, left=False,
    right=False, labelbottom=False, labelleft=False
)
plt.imshow(digits.images[0], cmap=plt.cm.gray_r)
```

```python
digits.target[0]
```

```python
fig, axes = plt.subplots(
    5, 10, figsize=(12, 7),
    subplot_kw={'xticks': [], 'yticks': []}
)
for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap=plt.cm.gray_r)
    ax.text(0.45, 1.05, str(digits.target[i]))
```

```python
plt.xticks([])
plt.hist(digits.target, rwidth=0.9)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    digits.data, digits.target, test_size=0.2, random_state=0
)
```

```python
model = LogisticRegression(max_iter=5000)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay as cmd
```

```python
import seaborn as sns
sns.set()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.grid(False)
cmd.from_estimator(model, x_test, y_test, cmap='Blues', colorbar=False, ax=ax)
```

```python
sns.reset_orig()
plt.tick_params(
    axis='both', which='both', bottom=False, top=False,
    left=False, right=False, labelbottom=False, labelleft=False
)
plt.imshow(digits.images[100], cmap=plt.cm.gray_r)
```
